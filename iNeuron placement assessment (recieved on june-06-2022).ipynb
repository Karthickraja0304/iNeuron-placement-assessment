```python
def find_and_replace(File_name):
    """Enter the filename with format in double quatations
       find_and_replace("abc.txt") will initiate the find and replace operation on abc Text Document"""
    try:
        Input_file=open(File_name,"r+")
        Updated_file=open("Updated_file.txt","w")
        find=input("Please enter the word to be searched = ")
        replace=input("Please enter the word to be replaced with = ")
        for lines in Input_file.readlines():
            if find in lines:
                updated_lines=lines.replace(find,replace)
                Updated_file.writelines(updated_lines)
                Updated_file.close()
                return("The updated file is saved as Updated.txt in your current working directory")
            else:
                return("The requested word is not available in the text file")
        Input_file.close()
    except Exception as e:
        print("We have encountered an error")
        return e 
```

```python
Abstract class:
        Abstract class is nothing but a blue print for a class yet to be created. It helps in sharing the creation 
        of a particular class among different programmers. Within Abstract class, there will be Abstract methods. 
        Abstract methods are nothing but methods which are titled but has no logic coded on to it. 
        Object creation is not possible in abstract class. Please refer below example,
    
```

```python
from abc import ABC, abstractmethod #importing ABSTRACT BASE CLASS and ABSTRACT METHOD
#This class is related to bike selection criteria for user purchase
class Bike(ABC): #inheritance
    @abstractmethod  #decorator
    def brand(self):
        pass
    def colour(self):
        pass
    def engine_cc(self):
        pass
    def mileage(self):
        pass
#The abstract methods are alloted to different programmers,where logic are coded seperateley and later loaded.
```

```python
Multiple inheritance:
        Multiple inheritence is a type of inheritance where two or more classes combined to give a child class
        which has  properties of all the combined classes similar to real life scenario wherein a child gets to 
        inherit both the traits of father and mother. Please refer below example,
```

```python
#This is a program for three machines as three classe with specific advantages, all are inherited into a single hybrid class

class machine1:
    def fuel_consumption(self):
        print("This machine is best for fuel efficiency")
class machine2:
    def space_occupied(self):
        print("This machine is known for its compact size")
class machine3:
    def noise_pollution(self):
        print("This machine exerts less noise")
class hybrid(machine1,machine2,machine3):#object created with this class will have access to all methods in all parent classes
    pass


new_engine=hybrid()  #Object creation of new_engine
new_engine.fuel_consumption()
new_engine.space_occupied()
new_engine.noise_pollution()     
#new_engine will have all advantages of those seperate three machines.
```

```python
Decorater:
    Decorator is a tool used in python which is used to modify any class or function, whose body is unknown to us.
    It helps to wrap a new function into a existing function(whose body is unknown) without affecting its nature.
```

```python
def func1(function): #function which takes other function as input
    def func2():# wrapping occurs
        print("Enters the func2")
        function() #input function 
        print("Ends here")
    return func2
@func1 #decorator used here adds the below function 
def user_function():
    print("execued with decorator")
    
user_function()
```
